```python
import random
import numpy as np
import scipy
from scipy.misc import derivative
import sympy as sp
import matplotlib.pyplot as plt
```

```python
x = sp.Symbol('x')
print(sp.diff(36/7*x**2+4,x))
```

```python
def f(x):
    return 36/7*x**2+4
```

```python
print(derivative(f,-8))
```

```python
alpha=0.001
Xold=-8
iterations = 45000
FunctionValues = []
FunctionValues.append(f(Xold))
Iterations = []
i = 0
Iterations.append(i)
while(i < iterations):
    
    Xnew=Xold-alpha*derivative(f,Xold)
    
    if (abs(f(Xold) - f(Xnew)) < 0.000000001):
        print("\n\n\nWe have finally minimized the function and come out of the loop")
        break
    
    Xold=Xnew
    
    i += 1
    FunctionValues.append(f(Xold))
    Iterations.append(i)

    print("The Value of Function at Xnew = {} is {}".format(Xnew,f(Xnew)))
```
